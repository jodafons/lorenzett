```python
def plot_layer(layer_data, norm = True):
  for t in range(layer_data.shape[2]):
    plt.imshow(layer_data[:,:,t],interpolation='nearest',aspect=layer_data.shape[1]/layer_data.shape[0])
    plt.title("Time sample: %d" % t)
    matplotlib.pyplot.gca().tick_params(axis='both', which='both', bottom='off', top='off',
                     left='off', right='off', labelbottom='off',
                     labelleft='off')
    plt.colorbar()
    plt.show()
```
